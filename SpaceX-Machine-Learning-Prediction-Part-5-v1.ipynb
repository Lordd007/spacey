This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

## TASK  12


Find the method performs best:


```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, classification_report, confusion_matrix
)

# ⬇️ List all the GridSearchCV objects you actually fit.
# Add/remove names as needed (keep the ('var_name', 'pretty name') pattern).
candidates = []
for var, name in [
    ('logreg_cv', 'Logistic Regression'),
    ('tree_cv',  'Decision Tree'),
    ('svm_cv',   'SVM'),
    ('knn_cv',   'KNN'),
]:
    if var in globals():
        candidates.append((name, globals()[var]))

assert candidates, "No fitted GridSearchCV objects found. Add your *_cv variables above."

# Helper to pick averaging for binary vs multi-class
binary = (len(np.unique(Y_test)) == 2)
avg = 'binary' if binary else 'weighted'

rows, reports, cms = [], {}, {}

for name, gscv in candidates:
    est = gscv.best_estimator_
    y_pred = est.predict(X_test)

    # ROC AUC if possible (binary only)
    roc = np.nan
    if binary:
        if hasattr(est, "predict_proba"):
            y_score = est.predict_proba(X_test)[:, 1]
            roc = roc_auc_score(Y_test, y_score)
        elif hasattr(est, "decision_function"):
            y_score = est.decision_function(X_test)
            roc = roc_auc_score(Y_test, y_score)

    rows.append({
        'model': name,
        'cv_mean_accuracy': gscv.best_score_,
        'test_accuracy': accuracy_score(Y_test, y_pred),
        'test_precision': precision_score(Y_test, y_pred, average=avg, zero_division=0),
        'test_recall': recall_score(Y_test, y_pred, average=avg, zero_division=0),
        'test_f1': f1_score(Y_test, y_pred, average=avg, zero_division=0),
        'test_roc_auc': roc
    })

    reports[name] = classification_report(Y_test, y_pred, zero_division=0)
    cms[name] = confusion_matrix(Y_test, y_pred)

summary = (pd.DataFrame(rows)
           .sort_values(by=['test_f1','test_accuracy'], ascending=False)
           .reset_index(drop=True))

print(summary)
best = summary.iloc[0]['model']
print(f"\nBest by F1: {best}\n")
print("Classification report:\n", reports[best])
print("Confusion matrix:\n", cms[best])
```

```python
df = summary[['model', 'test_accuracy']].sort_values('test_accuracy', ascending=False)

fig, ax = plt.subplots(figsize=(8,4))
bars = ax.bar(df['model'], df['test_accuracy'])

ax.set_ylabel('Accuracy')
ax.set_xlabel('Model')
ax.set_title('Model Comparison – Test Accuracy')
ax.set_ylim(0, 1)
plt.xticks(rotation=15, ha='right')

# value labels
ax.bar_label(bars, labels=[f"{v:.2f}" for v in df['test_accuracy']], padding=3, fontsize=9)

plt.tight_layout()
plt.savefig('model_test_accuracy.png', dpi=300, bbox_inches='tight')
plt.show()
```
